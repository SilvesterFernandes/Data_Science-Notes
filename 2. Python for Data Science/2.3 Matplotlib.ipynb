## **4.1 Line Plot**

- A line plot is one of the most basic types of plots used to **display data points** connected by **straight lines.**
- Its typically **used to visualize trends over time or continuous data.**
- In Matplotlib, a line plot can be **created using the plot() function.**

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-Axis (Horizantal)
y = [5, 10, 15, 20, 25] # Y-Axis (Verticle)

plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Creating the plot
plt.plot(x, y)
```

## **6. Subplots**
You can create multiple plots in one figure:

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)


fig, axs = plt.subplots(2, 2)  # 2 rows, 2 columns
axs[0, 0].plot(x, y)
axs[0, 1].bar(x, y)
axs[1, 0].scatter(x, y)
axs[1, 1].hist(y)

plt.tight_layout()  # Adjusts subplot parameters for better spacing
plt.show()
```

## **10. Interactivity :**
Matplotlib allows for some interactive features like zooming and panning, which can be useful when visualizing data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a figure and axis
fig, ax = plt.subplots()

# Plotting
ax.plot(x, y, label='Sine Wave', color='blue')

# Customizing
ax.set_title("Sine Function")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()

# Save the figure
plt.savefig('sine_wave.png')

# Show plot
plt.show()
```

**Adding Elements to Plots:**

**1 Adding Titles and Labels :**
- You can add titles to your plots and labels to your axes to make your visualizations clearer:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# Plotting
plt.plot(x, y)

# Adding title and axis labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the plot
plt.show()
```

**2 Adding Legends:**
- Legends help identify different datasets within a plot:

```python
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting two lines
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='red')

# Adding legend
plt.legend()

# Show the plot
plt.show()
```

**3.Adding Annotations:**
- You can highlight specific points using annotations.

```python
# Plotting
plt.plot(x, y1)

# Adding annotation at the peak of the sine wave
peak_x = np.pi / 2
peak_y = np.sin(peak_x)
plt.annotate('Peak', xy=(peak_x, peak_y), xytext=(peak_x + 1, peak_y + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Show the plot
plt.show()
```

**4.Combining Multiple Plots:**
- You can overlay multiple plots for comparison.

```python
# Plotting sine and cosine waves
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='red')

# Adding title, labels, and legend
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.show()
```

**Complete Example:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plotting
ax.plot(x, y1, label='Sine Wave', color='blue', linestyle='-')
ax.plot(x, y2, label='Cosine Wave', color='red', linestyle='--')

# Customizing
ax.set_title("Sine and Cosine Waves")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
ax.grid(True)

# Adding annotations
peak_x = np.pi / 2
peak_y = np.sin(peak_x)
ax.annotate('Peak of Sine', xy=(peak_x, peak_y), xytext=(peak_x + 1, peak_y + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Save the figure
plt.savefig('sine_cosine_waves.png')

# Show the plot
plt.show()
```
