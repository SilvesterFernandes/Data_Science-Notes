**Introduction:**

**2. Basic Usage:**

- Importing Matplotlib :

Typically, you will import pyplot from Matplotlib:

```python
import matplotlib.pyplot as plt
```

**3. Creating a Simple Plot:**
 
create a basic line plot:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-Axis (Horizantal)
y = [5, 10, 15, 20, 25] # Y-Axis (Verticle)

# Creating the plot
plt.plot(x, y)
```

**Adding titles and labels :**

```python
# Adding titles and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Displaying the plot
plt.show()
```

# **4. Plot Types**
Matplotlib supports various types of plots:

- **4.1 : Line Plot**
- **4.2 : Scatter plot**
- **4.3 : Bar Plot**
- **4.4 : Histogram**
- **4.5 : Pie Chart**
- **4.6 : Custom Plot**

## **4.1 Line Plot**

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-Axis (Horizantal)
y = [5, 10, 15, 20, 25] # Y-Axis (Verticle)

plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Creating the plot
plt.plot(x, y)
```

## **4.2 Scatter Plot :**

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5,6,7]            # X-Axis (Horizantal)
y = [2, 3, 7, 11, 10,18,9]         # Y-Axis (Verticle)

plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.scatter(x, y)
```

## **4.3 Bar Plot :**

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5,6,7]            # X-Axis (Horizantal)
y = [2, 3, 7, 11, 10,25,9]         # Y-Axis (Verticle)

plt.title("Bar Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.bar(x, y)
```

## **4.4 Histogram :**

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)

plt.title("Histogram")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.hist(y)
```

## **4.5 Pie Chart :**

```python
labels = ['Label_1', 'Label_2', 'Label_3']
sizes = [15, 30, 45]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```

## **5. Customizing Plots:**
You can customize plots with various options:

**5.1 Colors and Markers**

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)

plt.title("Custom ")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x, y, color='red', marker='o')
```

**5.2 Line Styles**

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)

plt.title("Dashed Line")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x, y, color="blue",linestyle='--')  # Dashed line
```

**5.3 Adding Legends :**

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)

plt.title("Legends")
plt.xlabel("X-Axis",color="Blue")
plt.ylabel("Y-Axis",color="Blue")
plt.plot(x, y, color="red",label='This is Line 1')
plt.legend()
```

**5.4 Annotations:**

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)

plt.title("Annotations")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.annotate('Max Value', xy=(x[4], y[4]), xytext=(x[4]+1, y[4]+5), arrowprops=dict(facecolor='black', shrink=0.05))
```

## **6. Subplots**
You can create multiple plots in one figure:

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)


fig, axs = plt.subplots(2, 2)  # 2 rows, 2 columns
axs[0, 0].plot(x, y)
axs[0, 1].bar(x, y)
axs[1, 0].scatter(x, y)
axs[1, 1].hist(y)

plt.tight_layout()  # Adjusts subplot parameters for better spacing
plt.show()
```

## **7. Saving Figures:**
To save a plot to a file, use:

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)

plot=plt.bar(x,y)
plt.savefig('my_plot.png')  # Save as PNG
```

## **8. Styling:**
Matplotlib offers several built-in styles to change the look of your plots:

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [1, 2, 3, 4, 5,6,7]         # Y-Axis (Verticle)
plot=plt.bar(x,y)
plt.style.use('ggplot')  # Use ggplot style
```

## **9. 3D Plots:**
For 3D plots, you need to import the mpl_toolkits.mplot3d module:

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Sample data
x = [2, 4, 6, 8, 10,12,14]            # X-Axis (Horizantal)
y = [10, 25, 34, 40, 55,69,72]         # Y-Axis (Verticle)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, zs=0)
```

Example Code Snippet
Heres an example that encompasses many of the concepts:


import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a figure and axis
fig, ax = plt.subplots()

# Plotting
ax.plot(x, y, label='Sine Wave', color='blue')

# Customizing
ax.set_title("Sine Function")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()

# Save the figure
plt.savefig('sine_wave.png')

# Show plot
plt.show()
